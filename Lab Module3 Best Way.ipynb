```python
# Applying the concepts of input functions, list and list methods, break and continue,
# conditional and loop statement.

# Challenge #1. Create a Python program that enters and displays the equipment name entered.

# Continually prompt the user to enter an equipment item until they entered a blank item(empty). 
# After all the items are entered, display the equipment list back to the user.

# Provide your code below:

Equipment_List = []

while True:
    device = input("Enter the device hostname.\n")
    if device == "":
        break
    Equipment_List.append(device)
    print(f"{Equipment_List[-1]} Device added.\n")

print("Equipment List:")
print("-"*15)

for item in Equipment_List:
    print(item)


    


```
